138. [Copy List with Random Pointer](https://leetcode.com/problems/copy-list-with-random-pointer/description/)

construct list with new node next to original to ensure fast access for random pointer

1->1'->2->2'->3->3'

````python
"""
# Definition for a Node.
class Node:
    def __init__(self, x: int, next: 'Node' = None, random: 'Node' = None):
        self.val = int(x)
        self.next = next
        self.random = random
"""


class Solution:
    def copyRandomList(self, head: "Optional[Node]") -> "Optional[Node]":
        if head is None:
            return head
        # add new node next to original
        curr = head
        while curr is not None:
            tmp = curr.next
            node = Node(curr.val)
            curr.next = node
            node.next = tmp
            curr = tmp
        # link random
        curr = head
        while curr is not None:
            curr.next.random = curr.random.next if curr.random else None
            curr = curr.next.next
        curr = head
        ptr = head.next
        ret = head.next
        # link next and restore old list
        while curr:
            curr.next = ptr.next
            ptr.next = ptr.next.next if ptr.next else None
            curr = curr.next
            ptr = ptr.next

        return ret
````

142. [Linked List Cycle II](https://leetcode.com/problems/linked-list-cycle-ii/description/)

$a+b=k * c$ to deduce $a = k * c - b$ to find the entrance for cycle


````python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None


class Solution:
    def detectCycle(self, head: Optional[ListNode]) -> Optional[ListNode]:
        slow = head
        fast = head
        while fast is not None and fast.next is not None:
            slow = slow.next
            fast = fast.next.next
            if slow == fast:
                break

        if fast is None or fast.next is None:
            return None

        slow = head
        while slow != fast:
            slow = slow.next
            fast = fast.next
        return slow
````

````python
from queue import Queue

class MyStack:

    def __init__(self):
        self.stack = Queue()

    def push(self, x: int) -> None:
        sz = self.stack.qsize()
        self.stack.put(x)
        for i in range(sz):
            self.stack.put(self.stack.get())

    def pop(self) -> int:
        return self.stack.get()

    def top(self) -> int:
        return self.stack.queue[0]

    def empty(self) -> bool:
        return self.stack.empty()
````

````python
def minPrice(a: list[int],p:list[int])->int:

  def process1(i,curr_a):
    if i >= len(a):
      return 0
    if curr_a < a[i]:
      return p[i] + process1(i+1,curr_a + a[i])
    else:
      return min(p[i] + process1(i+1,curr_a + a[i]), process1(i+1,curr_a))

  #process1(1,p[0],a[0])

  def process2()->int:
    n = len(a)
    m = sum(a);
    dp = [[0 for _ in range(m+1)] for _ in range(n+1)]
    curr_a = 0
    # dp: min price required at i, j
    for i in range(n-1,-1,-1):
        for j in range(m+1):
          if j + a[i] > m :
            continue
          if j < a[i]:
            dp[i][j] = p[i] + dp[i+1][j+a[i]]
          else:
            dp[i][j] = min(p[i] + dp[i+1][j+a[i]], dp[i+1][j])
    #print(dp)

    return dp[0][0]

  def process3() -> int:
    n = len(a)
    m = sum(p) + 1;
    # [n][m+1]
    dp = [[0 for _ in range(m)] for _ in range(n)]
    for i in range(n):
      for j in range(m):
        dp[i][j] = -1
    # buy first ability
    dp[0][p[0]] = a[0]
    #print(dp)
    for i in range(1,n):
      for j in range(1,m):
        if j > p[i] and dp[i-1][j-p[i]] != -1:
          dp[i][j] = a[i] + dp[i-1][j-p[i]]
        if dp[i-1][j] >= a[i]:
          dp[i][j] = max(dp[i][j],dp[i-1][j])
    #print(dp)
    for j in range(m):
      if dp[n-1][j] != -1:
        return j

    return -1

  return process3()



minPrice([7,2,9,4,8],[3,6,1,5,10])
````

### ***Use Recursion To Sort A Stack***
```
input: [5,4,5,4,3,6,3,6,1]
output:[6,6,5,5,4,4,3,3,1]
```

````python
class sort_stack:

  def sort(self,stack:list[int])->list[int]:
    depth = self.find_depth(stack)
    #print(f"curr depth: {depth}, {len(stack)}")
    while depth > 0:
      curr = self.find_max(stack, depth)
      occurance = self.find_occurance(stack,depth,curr)
      self.down(stack,depth,curr,occurance)
      depth -= occurance
    return stack

  def find_depth(self, stack:list[int])->int:
    if not stack:
      return 0
    ele = stack.pop()
    depth = 1 + self.find_depth(stack)
    stack.append(ele)
    return depth

  def find_max(self,stack:list[int],depth:int)->int:
    if depth == 0:
      return float('-inf')
    curr = stack.pop()
    curr_max = max(curr,self.find_max(stack,depth - 1))
    stack.append(curr)
    return curr_max

  def find_occurance(self, stack:list[int],depth:int,max_ele:int)->int:
    if depth == 0:
      return 0
    curr = stack.pop()
    occ = 0
    occ += (1 if curr == max_ele else 0)
    occ += self.find_occurance(stack,depth - 1, max_ele)
    stack.append(curr)
    return occ

  def down(self, stack:list[int],depth:int,ele:int,occ:int):
    if depth == 0:
      for _ in range(occ):
        stack.append(ele)
      return
    curr = stack.pop()
    self.down(stack,depth - 1, ele, occ)
    if curr != ele:
      stack.append(curr)

sl = sort_stack()
print(sl.sort([5,4,5,4,3,6,3,6,1]))
````

🎯 Goal:

📌 Constraints:\

````python
"""
monotonic: more rounds accured monster will have less hp
range: 1 ~ hp + 1
f:
max(cut[i], (round - i) * poison)

"""
import sys


class Solution:

  def kill_monster(self,cuts:list[int],poisons:list[int], hp: int)->int:
    l, r = 1, hp + 1
    ans = 0
    while l <= r:
      mid = (l+r) // 2
      if self.f(mid, cuts, poisons, hp):
        ans = mid
        r = mid - 1
      else:
        l = mid + 1
    return ans

  def f(target, cuts:list[int],poisons:list[int], hp: int)->bool:
    n = min(len(cuts),target)
    curr = 0
    for i in range(n):
      j = i + 1
      hp -= max(cuts[i],poisons[i]*(target - j))
      if hp <= 0: return True

    return False


class Codec:

  def fast1(self,cuts, poisons, hp):
    total_poison = sum(poisons)
    n = len(cuts)
    # Initialize 3D DP array with 0
    dp = [[[0 for _ in range(total_poison + 1)] for _ in range(hp + 1)] for _ in range(n)]
    return self.f1(cuts, poisons, 0, hp, 0, dp)

  def f1(self,cuts, poisons, i, r, p, dp):
    r -= p
    if r <= 0:
        return i + 1
    if i == len(cuts):
        if p == 0:
            return sys.maxsize  # Equivalent to Integer.MAX_VALUE
        else:
            return len(cuts) + 1 + (r + p - 1) // p  # Integer division with ceiling
    if dp[i][r][p] != 0:
        return dp[i][r][p]

    # Option 1: use cut[i] if it helps
    if r <= cuts[i]:
        p1 = i + 1
    else:
        p1 = self.f1(cuts, poisons, i + 1, r - cuts[i], p, dp)

    # Option 2: use poison[i]
    p2 = self.f1(cuts, poisons, i + 1, r, p + poisons[i], dp)

    ans = min(p1, p2)
    dp[i][r][p] = ans
    return ans
````

Example 1:

Thinking:

- base case: when nothing to compare answer is 0 \
  "" abc -> 0  or abd "" -> 0
- when character is matched s1[i] == s2[j] they has to match bc it's expanding the answer
- when no match either proceed with s1[i-1] or s2[j-1]

````python
class Solution:
    def longestCommonSubsequence(self, text1: str, text2: str) -> int:
      n, m = len(text1), len(text2)
      if n < m:
        return self.longestCommonSubsequence(text2,text1)
      dp = [0]*(m+1)
      for i in range(1,n+1):
        left, backup = 0, 0
        for j in range(1,m+1):
          backup = dp[j]
          # update dp[j]
          if text1[i-1] == text2[j-1]:
            dp[j] = 1 + left
          else:
            dp[j] = max(dp[j],dp[j-1])
          left = backup
      return dp[m]

def run_tests():
    tests = [
        (("abcde", "ace"), 3, 'Example 1'),
        (("abc",   "abc"), 3, 'Example 2'),
        (("abc",   "def"), 0, 'Example 3'),
    ]
    sol = Solution()
    passed = 0
    for (t1, t2), expected, name in tests:
        got = sol.longestCommonSubsequence(t1, t2)
        ok = "PASS" if got == expected else "FAIL"
        print(f"{ok}: {name}: got={got}, expected={expected} (text1={t1}, text2={t2})")
        passed += ok == "PASS"
    print(f"\n{passed}/{len(tests)} tests passed.")

if __name__ == "__main__":
    run_tests()
````

Start dp with base case. Singe length letter is palindrome itself:\

now the general dp, we need to find the shorter length answer for $f$ to solve $f(l,r)$

$max(f(l+1,r),f(l,r-1))$ \

to visualize the dp table: \
$$
\begin{bmatrix}
1 & 0 & 0 & 0 & 0 \\
x & 1 & 0 & 0 & 0 \\
x & x & 1 & 0 & 0 \\
x & x & x & 1 & 0 \\
x & x & x & x & 1
\end{bmatrix}
$$
the answer we need to fill is the zeros because $l \leq r$


````python
class Solution:
    def longestPalindromeSubseq(self, s: str) -> int:
      # return self.f2(s)
      return self.f3(s)

    def f3(self, s:str)->int:
      n = len(s)
      if n == 0:
        return 0

      dp = [0]*n
      for l in range(n-1,-1,-1):
        dp[l] = 1
        if l + 1 < n:
          dp[l+1] = 2 if s[l] == s[l+1] else 1
        diag = dp[l]
        for r in range(l+2,n):
          backup = dp[r]
          if s[l] == s[r]:
            dp[r] = 2 + diag
          else:
            dp[r] = max(dp[r],dp[r-1])
          diag = backup
      return dp[n-1]


    def f2(self, s: str)->int:
        n = len(s)
        if n == 0:
            return 0

        dp = [[0]*n for _ in range(n)]

        for l in range(n-1, -1, -1):
            dp[l][l] = 1
            if l + 1 < n:
                dp[l][l+1] = 2 if s[l] == s[l+1] else 1
            for r in range(l+2, n):
                if s[l] == s[r]:
                    dp[l][r] = 2 + dp[l+1][r-1]
                else:
                    dp[l][r] = max(dp[l+1][r], dp[l][r-1])

        return dp[0][n-1]

def run_tests():
    tests = [
        ("bbbab", 4, 'Example 1'),
        ("cbbd", 2, 'Example 2'),
    ]
    sol = Solution()
    passed = 0
    for s, expected, name in tests:
        got = sol.longestPalindromeSubseq(s)
        ok = "PASS" if got == expected else "FAIL"
        print(f"{ok}: {name}: got={got}, expected={expected} (s={s})")
        passed += ok == "PASS"
    print(f"\n{passed}/{len(tests)} tests passed.")

if __name__ == "__main__":
    run_tests()
````

````python
class Solution:
  def generate_binary_tree(self,n:int, m:int)->int:
    dp = [[-1]*(m) for _ in range(n)]
    return self.f1(n,m, dp)

  def f1(self, n:int, m:int, dp:list[list[int]])->int:
    if n == 0:
      return 1
    if m == 0:
      return 0

    if dp[n-1][m-1] > -1:
      return dp[n-1][m-1]

    ans = 0
    for i in range(n):
      ans += self.f1(i, m - 1,dp) * self.f1(n - i - 1, m - 1,dp)
    dp[n-1][m-1] = ans;
    return ans

def run_tests():
    tests = [
        (3, 3, 5, 'Example 1'),
        (3, 2, 1, 'Example 2'),
        (4, 3, 6, 'Example 3'),
    ]
    sol = Solution()
    passed = 0
    for i1, i2, expected, name in tests:
        got = sol.generate_binary_tree(i1,i2)
        ok = "PASS" if got == expected else "FAIL"
        print(f"{ok}: {name}: got={got}, expected={expected} (n={i1} m={i2})")
        passed += ok == "PASS"
    print(f"\n{passed}/{len(tests)} tests passed.")

if __name__ == "__main__":
    run_tests()
````

Effective Coloring Problem

Given: Two positive integers n, m (1 ≤ n, m ≤ 5000).

Description: There is a sequence of $n$ cells. Each cell can be painted in one color chosen from $m$ available colors. A coloring is effective if all $n$ cells are painted and each of the $m$ colors appears at least once.

Task: Compute the total number of effective colorings and output the result modulo $10^7$


### Think

dp[i][j]: given i cells and j color "needed" how many valid coloring

base case:   
m = 0 no coloring avaliable dp[i][0] is all 0 \
m = 1 dp[i][1] = m --- need one color we can pick any one from m colors

dp[i][j]:
1. dp[i-1][j] used all m colors then we just need to pick any color from m

    dp[i][j] = dp[i-1][j] * j

2. dp[i-1][j] did not pick all colors we need to pick any color from m - j + 1

    dp[i][j] = dp[i-1][j-1] * (m-j+1) + dp[i][j]

````python
class Solution:
  MOD = 1000000007
  def validCellColor(self,n:int, m:int)->int:
    return self.f1(n,m);

  def f1(self, n:int, m:int)->int:
    if m > n:
      return 0

    dp = [[0]*(m+1) for _ in range(n+1)]
    dp[0][0] = 1
    for i in range(1,n+1):
      dp[i][1] = m

    for i in range(2,n+1):
      for j in range(2,m+1):
        dp[i][j] = (dp[i-1][j] * j) % self.MOD
        dp[i][j] = (dp[i-1][j-1] * (m - j + 1)+dp[i][j]) % self.MOD

    return dp[n][m]


class Comparator:
  MOD = 1_000_000_007
  def validCellCompare(self, n: int, m: int) -> int:
        if m > n:
            return 0
        path = [0] * n
        used = [False] * (m + 1)  # 1..m
        return self._dfs(path, used, 0, n, m, 0) % self.MOD

  def _dfs(self, path, used, i, n, m, used_cnt):
      # prune: even if we use a new color at every remaining position,
      # we can't reach m distinct colors
      if used_cnt + (n - i) < m:
          return 0

      if i == n:
          return 1 if used_cnt == m else 0

      total = 0
      for color in range(1, m + 1):
          was = used[color]
          if not was:
              used[color] = True
              path[i] = color
              total += self._dfs(path, used, i + 1, n, m, used_cnt + 1)
              used[color] = False
          else:
              path[i] = color
              total += self._dfs(path, used, i + 1, n, m, used_cnt)
      return total % self.MOD

def run_tests():

  sol = Solution()
  comp = Comparator()

  end = 6
  for i in range(1,end):
    for j in range(1,end):
      ans_1 = sol.validCellColor(i,j)
      ans_2 = comp.validCellCompare(i,j)
      ok = "PASS" if ans_1 == ans_2 else "FAIL"
      print(f"\n{ok}: n={i},m={j} answer={ans_1} compare={ans_2}")


if __name__ == "__main__":
    run_tests()
````

#### Delete Minimum Characters to Become a Substring

Given two strings s1 and s2, return the minimum number of characters that must be deleted from s1 so that the resulting string becomes a substring of s2.

```
inputs: s1="abcdf" s2="acbdecjgabcfeze"
output: 1
```

```
inputs: s1="aaaa" s2="ba"
output: 3
```

#### Think

dp[i][j] at index i from s1[0...i-1] and index j s2[0...j-1] min number of character to delete so that s1 suffix of s2.

The answer will be the min value at the last row.

dp[i][j]:

1. s1[i-1] $=$ s2[j-1] -> dp[i][j] = dp[i-1][j-1] \
when equals nothing to delete, need to compare previous characters
2. s1[i-1] $\neq$ s2[i-1] -> dp[i][j] = 1 + dp[i-1][j] \
when not equal, need to delete s1[i-1] and find a match from s[i-2]

base case:
dp[i][0] = i since no character in s2 we need to delete all characters




````python
class Solution:
  def delete_min_char(self,s1:str,s2:str)->int:
    return self.f1(s1,s2)


  def f1(self, s1:str, s2:str)->int:
    n, m = len(s1), len(s2)

    dp = [[0] * (m+1) for _ in range(n+1)]

    for i in range(1,n+1):
      dp[i][0] = i
      for j in range(1,m+1):
        if s1[i-1] == s2[j-1]:
          dp[i][j] = dp[i-1][j-1]
        else:
          dp[i][j] = 1 + dp[i-1][j]
    return min(dp[n])


class Compare:
  """
  generate all subsequence of s1 to compare with s2
  if there is a substring ans = len(s1) - len(seq)
  """
  def delete_min_char(self, s1: str, s2: str) -> int:
      # recompute per call, no leakage
      seq_list = []
      self._gen(0, "", s1, seq_list)
      # longest first
      seq_list.sort(key=len, reverse=True)
      for seq in seq_list:
          if s2.find(seq) != -1:         # substring check
              return len(s1) - len(seq)  # deletions
      return len(s1)

  def _gen(self, i: int, cur: str, s: str, out: list[str]):
      if i == len(s):
          out.append(cur)
          return
      self._gen(i + 1, cur, s, out)            # skip s[i]
      self._gen(i + 1, cur + s[i], s, out)     # take s[i]


import random, string

def rand_lower(n: int) -> str:
    return ''.join(random.choices(string.ascii_lowercase, k=n))

def run_tests():
  sol = Solution()
  print("Test starts")
  for _ in range(2000):
    comp = Compare()
    s1 = rand_lower(random.randint(1,7))
    s2 = rand_lower(random.randint(1,7))
    ans1 = sol.delete_min_char(s1,s2)
    ans2 = comp.delete_min_char(s1,s2)
    if ans1 != ans2:
      print(f"FAILED with s1={s1} s2={s2}")
      print(f"ans1={ans1} ans2={ans2}")
      return

  print("PASS")


if __name__ == "__main__":
    run_tests()
````

## Subarray Sum DP

### [53. Max Subarray Sum](https://leetcode.com/problems/maximum-subarray/)

extention question:
find maximum subarray sum'ss left and right index and it's sum


### Think

l - current left index
r - current right index
prev - current subarray sum

there are two cases $prev \geq 0$ and $prev < 0$
- if $pre \geq 0$ then need to include previous sum and update l, r
- $pre < 0$ then do not include previous sum start from $prev = num[i]$, $l=r=i$

````python
from typing import Tuple
class Solution:

  def find_max_subarray_info(self,nums:list[int])->tuple[int,int,int]:
    l,r,prev = 0,0,nums[0]
    left, right, sum = l,r,prev

    for i in range(1,len(nums)):
      if prev >= 0:
        prev = prev + nums[i]
      else:
        prev = nums[i]
        l = i
      r = i
      if prev > sum:
        left, right, sum = l,r,prev


    return (left,right,sum)




class Compare:

  def _find_max_subarray_info(self,nums:list[int])->tuple[int,int,int]:
    left, right, sum = 0,0,nums[0]

    for i in range(len(nums)):
      l,r,curr = i,i,nums[i]
      if curr > sum:
          left,right,sum = l,r,curr
      for j in range(i+1,len(nums)):
        curr = curr + nums[j]
        if curr > sum:
          left,right,sum = i,j,curr


    return (left,right,sum)

import random

# n integers in [low, high] inclusive
def rand_int_array(n, low, high=20):
    return [random.randint(low, high) for _ in range(n)]

def run_tests():

  sol = Solution()
  cp = Compare()
  n = 2000
  for _ in range(n):
    nums = rand_int_array(random.randint(1,20),random.randint(-10,10),)
    s1=sol.find_max_subarray_info(nums)
    s2 = cp._find_max_subarray_info(nums)
    if s1 != s2:
      print(f'Failed with nums={nums} sol={s1} cp={s2}')


  print("PASS")


if __name__ == "__main__":
    run_tests()




````

### Max Submatrix Sum

Given a n * m matrix, find the max submatrix sum and return upper left indx and lower right index [r1,c1,r2,c2]

input:
[ \
   [-1,0],\
   [0,-1]\
]

output: [0,1,0,1]


1 <= matrix.length, matrix[0].length <= 200

````python
class Solution:
  def max_submatrix_sum(self, grid:list[list[int]])->list[int]:
    n = len(grid)
    m = len(grid[0])
    curr = [0]*m
    r1,c1,r2,c2 = 0,0,0,0
    ans = float('-inf')
    for up in range(n):
      curr = [0] * m
      for down in range(up,n):
        for c in range(m):
          curr[c] += grid[down][c]

        #1d find max subarray sum
        l, prev = 0,0
        for i,v in enumerate(curr):
          if prev >= 0:
            prev += v
          else:
            prev = v
            l = i

          if prev > ans:
            ans = prev
            r1,c1,r2,c2 = up,l,down,i

    return [r1,c1,r2,c2]


class Compare:
  def max_submatrix_sum_compare(self, grid:list[list[int]])->list[int]:
    n = len(grid)
    m = len(grid[0])
    r1,c1,r2,c2 = 0,0,0,0
    ans = float('-inf')
    for up in range(n):
      for down in range(up,n):
        for left in range(m):
          for right in range(left,m):
            curr = 0
            for r in range(up,down+1):
              for c in range(left,right+1):
                curr += grid[r][c]

            if curr > ans:
              r1 = up
              c1 = left
              r2 = down
              c2 = right
              ans = curr
    #print(ans)
    return [r1,c1,r2,c2]

import random


# n integers in [low, high] inclusive
def rand_int_array(n, low, high=20):
    return [random.randint(low, high) for _ in range(n)]

def run_tests():

  sl = Solution()
  cp = Compare()
  cnt = 200
  for _ in range(cnt):
    n = random.randint(1,20)
    m = random.randint(1,20)
    grid = []
    #print("hello")
    for _ in range(n):
      grid.append(rand_int_array(m,-10,10))
      a1 = sl.max_submatrix_sum(grid)
      a2 = cp.max_submatrix_sum_compare(grid)
      if a1 != a2:
        print(f'grid={grid}')
        print("WRONG ANSER")
        print(f'solution={a1} compare={a2}')
        return
  print("PASS")


if __name__ == "__main__":
  run_tests()



````

### Magic Scroll

Given a nums integer array, a magic scroll can change numbers in the array nums to 0 from l to r where 0 <= l <= r < nums.length. Find the maximum array's sum with at most 2 magic scroll use.

```
input: [1,-1,-3,4,5,6,0]
output: 16
with 0 scroll use maximum subarray sum is [4,5,6,0]

input:[-1,-2,-3,-4]
output: 0

with 1 scroll use to change numbers from 0 to 3 -> [0,0,0,0]
```

0<= nums.length <= 10^5



````python
class Solution:
  def magic_scroll(self,nums:list[int])->int:
    n = len(nums)

    #1. no scroll used
    p1 = 0
    for v in nums:
      p1 = p1 + v

    # 2. one scroll
    # prefix[i] -> must use one scroll the max sum can obtain from 0..i-1
    # eithre scroll removed i or not
    prefix = [0] * n
    sum = nums[0]
    max_prefix = max(0,nums[0])
    for i in range(1,n):
      prefix[i] = max(prefix[i-1] + nums[i],max_prefix)
      sum = sum + nums[i]
      max_prefix = max(max_prefix,sum)
    p2 = prefix[n-1]

    # suffix[i] -> must use one scroll the max sum can obtain from i..n-1
    suffix = [0]*n
    sum = nums[n-1]
    max_suffix = max(0,nums[n-1])
    for i in range(n-2,-1,-1):
      suffix[i] = max(suffix[i+1] + nums[i],max_suffix)
      sum = sum + nums[i]
      max_suffix = max(max_suffix,sum)
    #3.use two scroll, if index i is not covered meaning that we used scroll at some 0..i-1 position and i+1..n-1 position
    # otherwise if covered i then it's just scenario one with overlap intervals
    p3 = float('-inf')
    for i in range(1,n):
      p3 = max(prefix[i-1] + suffix[i], p3)

    return max(p3,max(p1,p2))


class Compare:
  def magic_scroll(self,nums:list[int])->int:
    n = len(nums)
    p1 = 0
    sum = 0
    for v in nums:
      p1 = p1 + v
    p2 = self.f1(0,n-1,nums)
    p3 = float('-inf')
    for i in range(1,n):
      p3 = max(p3, self.f1(0,i-1,nums) + self.f1(i,n-1,nums))
    return max(p1,max(p2,p3))

  def f1(self,l:int, r:int,nums:list[int])->int:
    ans = float('-inf')
    for a in range(l,r+1):
      for b in range(a,r+1):
        curr = 0;
        for i in range(l,a):
          curr = curr + nums[i]
        for j in range(b+1,r+1):
          curr = curr + nums[j]
        ans = max(ans,curr)
    return ans

import random
# n integers in [low, high] inclusive
def rand_int_array(n, low, high=20):
    return [random.randint(low, high) for _ in range(n)]

def run_tests():
  sl = Solution()
  cp = Compare()
  n = 1000
  for _ in range(n):
    nums = rand_int_array(random.randint(1,10),random.randint(-10,0),random.randint(1,10))
    a1 = sl.magic_scroll(nums)
    a2 = cp.magic_scroll(nums)
    if a1 != a2:
      print("WRONG!!!")
      print(f'nums={nums} a1={a1} a2={a2}')
      return
  print("PASS")


if __name__ == "__main__":
  run_tests()
````
