You have **an operation** to choose i'th element or not to choose it. Usually it's looking for at index i we are looking for some answer on condition j(target sum most likely)

We need to find a way to transform the probelm to the model problem, "Given int array, find if they can come up with the target sum \
dp[i][j] -> \
not choose i: dp[i-1][j] \
choose i: dp[i-1][j-nums[i]] \

dp[0][0] always has answer since it's a empty set **true by default** 

```python
for num in nums:
    for j in range(m,-1,-1):
        dp[j] += dp[j-num];
```
